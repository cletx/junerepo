```python
#Write a function that takes a list of numbers and returns a list containing the square of all even numbers
def square_even_numbers(numbers):
    return [x**2 for x in numbers if x % 2 == 0]
nums = [1, 2, 3, 4, 5, 6]
result = square_even_numbers(nums)
print(result)  # Output: [4, 16, 36]
```

```python
#Create a function is_palindrome() that checks if a string is a palindrome (same forward and backward) using basic string operations
def is_palindrome(user_input):
  print(user_input)
  if user_input == user_input[::-1]:
    print("is palindrome")
  else:
    print("not a palindrome")
is_palindrome("television")

is_palindrome("plumber")
is_palindrome("printer")
```

```python
#define a function that takes two parameters and returns their product. Then, define the same logic using a lambda expression and call it.
def multiply(a, b):
    return a * b
result1 = multiply(5, 6)
print("Regular function result:", result1)

# Lambda expression
multiply_lambda = lambda a, b: a * b
result2 = multiply_lambda(5, 6)
print("Lambda function result:", result2)  # Output: 30
```

```python
#Create a function that accepts a string and returns the number of vowels in it.
def count_vowels(s):
    vowels = 'aeiouAEIOU'
    return sum(1 for char in s if char in vowels)
text = "Hello, World!"
result = count_vowels(text)
print("Number of vowels:", result)  # Output: 3
```

```python
#Write a nested function where the outer function has a variable and the inner function modifies it using the nonlocal keyword.
def outer_function():
  outer_variable = "hello outer"
  def inner_function():
    nonlocal outer_variable
    outer_variable= "hello from inside"
    print(outer_variable)
  print("before calling inner function",outer_variable)
  inner_function()
  print("after calling inner function",outer_variable)

```

```python
#Use a global variable to store the count of how many times a function is called. Print the counter each time the function runs.
# Global variable to store the count
call_count = 0

def my_function():
    global call_count
    call_count += 1
    print(f"Function has been called {call_count} times.")

my_function()
my_function()
my_function()
```
